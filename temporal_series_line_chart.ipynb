```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
data = pd.read_csv('/content/data4.csv')
data.head()
```

```python
data.dtypes
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
data.head()
```

```python
data.dtypes
```

```python
data.shape
```

**Examlpe One**

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(12, 6))

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot(dates, y, linestyle='solid', marker='o')
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y') #month, day, year
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
#dates = [datetime.strptime(d, '%Y-%m-%d') for d in data['date']]

plt.show()
```

**Example Two**

```python
plt.style.use('ggplot')

plt.figure(figsize=(12, 6))

price_date = data['Date']
price_close = data['Close']

plt.plot(price_date, price_close, linestyle='solid', marker='o')
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y') #month, day, year
plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
