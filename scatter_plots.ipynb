```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/content/2019-05-31-data.csv')
data.head()
```

```python
data['views_likes_ratio'] = data['view_count'] / data['likes']
data.head()
```

Example one

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(12, 6))

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

Example Two

```python
plt.style.use('ggplot')

plt.figure(figsize=(12, 6))

view_count = data['view_count']
likes = data['likes']
views_likes_ratio = data['views_likes_ratio']

plt.scatter(view_count, likes, c=views_likes_ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Views/Likes Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```
